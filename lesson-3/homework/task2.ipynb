```python
# 1. Count Occurrences
def count_occurrences(tpl, element):
    return tpl.count(element)

# 2. Max Element
def max_element(tpl):
    return max(tpl)

# 3. Min Element
def min_element(tpl):
    return min(tpl)

# 4. Check Element
def check_element(tpl, element):
    return element in tpl

# 5. First Element
def first_element(tpl):
    return tpl[0] if tpl else None

# 6. Last Element
def last_element(tpl):
    return tpl[-1] if tpl else None

# 7. Tuple Length
def tuple_length(tpl):
    return len(tpl)

# 8. Slice Tuple
def slice_tuple(tpl):
    return tpl[:3]

# 9. Concatenate Tuples
def concatenate_tuples(tpl1, tpl2):
    return tpl1 + tpl2

# 10. Check if Tuple is Empty
def is_tuple_empty(tpl):
    return len(tpl) == 0

# 11. Get All Indices of Element
def get_all_indices(tpl, element):
    return [i for i, value in enumerate(tpl) if value == element]

# 12. Find Second Largest
def second_largest(tpl):
    unique_tpl = tuple(set(tpl))  # Remove duplicates
    sorted_tpl = sorted(unique_tpl, reverse=True)
    return sorted_tpl[1] if len(sorted_tpl) > 1 else None

# 13. Find Second Smallest
def second_smallest(tpl):
    unique_tpl = tuple(set(tpl))  # Remove duplicates
    sorted_tpl = sorted(unique_tpl)
    return sorted_tpl[1] if len(sorted_tpl) > 1 else None

# 14. Create a Single Element Tuple
def single_element_tuple(element):
    return (element,)

# 15. Convert List to Tuple
def list_to_tuple(lst):
    return tuple(lst)

# 16. Check if Tuple is Sorted
def is_tuple_sorted(tpl):
    return tpl == tuple(sorted(tpl))

# 17. Find Maximum of Subtuple
def max_of_subtuple(tpl, start, end):
    return max(tpl[start:end])

# 18. Find Minimum of Subtuple
def min_of_subtuple(tpl, start, end):
    return min(tpl[start:end])

# 19. Remove Element by Value
def remove_element(tpl, element):
    lst = list(tpl)
    if element in lst:
        lst.remove(element)
    return tuple(lst)

# 20. Create Nested Tuple
def create_nested_tuple(tpl, size):
    return tuple(tpl[i:i + size] for i in range(0, len(tpl), size))

# 21. Repeat Elements
def repeat_elements(tpl, times):
    return tuple(value for value in tpl for _ in range(times))

# 22. Create Range Tuple
def create_range_tuple(start, end):
    return tuple(range(start, end + 1))

# 23. Reverse Tuple
def reverse_tuple(tpl):
    return tpl[::-1]

# 24. Check Palindrome
def check_palindrome(tpl):
    return tpl == tpl[::-1]

# 25. Get Unique Elements
def unique_elements(tpl):
    seen = set()
    return tuple(x for x in tpl if not (x in seen or seen.add(x)))

# Examples for testing:
example_tuple = (1, 2, 3, 2, 4, 1, 5)
print("Occurrences of 2:", count_occurrences(example_tuple, 2))
print("Max element:", max_element(example_tuple))
print("Min element:", min_element(example_tuple))
print("Check element 3 in tuple:", check_element(example_tuple, 3))
print("First element:", first_element(example_tuple))
print("Last element:", last_element(example_tuple))
print("Tuple length:", tuple_length(example_tuple))
print("Slice tuple:", slice_tuple(example_tuple))
print("Concatenate tuples:", concatenate_tuples((1, 2), (3, 4)))
print("Is tuple empty:", is_tuple_empty(()))
print("Indices of 2:", get_all_indices(example_tuple, 2))
print("Second largest element:", second_largest(example_tuple))
print("Second smallest element:", second_smallest(example_tuple))
print("Single element tuple:", single_element_tuple(42))
print("Convert list to tuple:", list_to_tuple([1, 2, 3]))
print("Is tuple sorted:", is_tuple_sorted((1, 2, 3)))
print("Max of subtuple:", max_of_subtuple(example_tuple, 1, 4))
print("Min of subtuple:", min_of_subtuple(example_tuple, 1, 4))
print("Remove element 2:", remove_element(example_tuple, 2))
print("Create nested tuple:", create_nested_tuple(example_tuple, 2))
print("Repeat elements 3 times:", repeat_elements((1, 2), 3))
print("Range tuple:", create_range_tuple(1, 5))
print("Reverse tuple:", reverse_tuple(example_tuple))
print("Check palindrome:", check_palindrome((1, 2, 3, 2, 1)))
print("Unique elements:", unique_elements(example_tuple))
```
