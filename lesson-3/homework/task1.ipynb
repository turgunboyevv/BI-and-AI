```python
# Program 1: Count Occurrences
def count_occurrences(lst, element):
    return lst.count(element)

# Program 2: Sum of Elements
def sum_of_elements(lst):
    return sum(lst)

# Program 3: Max Element
def max_element(lst):
    return max(lst)

# Program 4: Min Element
def min_element(lst):
    return min(lst)

# Program 5: Check Element
def check_element(lst, element):
    return element in lst

# Program 6: First Element
def first_element(lst):
    return lst[0] if lst else None

# Program 7: Last Element
def last_element(lst):
    return lst[-1] if lst else None

# Program 8: Slice List
def slice_list(lst):
    return lst[:3]

# Program 9: Reverse List
def reverse_list(lst):
    return lst[::-1]

# Program 10: Sort List
def sort_list(lst):
    return sorted(lst)

# Program 11: Remove Duplicates
def remove_duplicates(lst):
    return list(set(lst))

# Program 12: Insert Element
def insert_element(lst, index, element):
    lst.insert(index, element)
    return lst

# Program 13: Index of Element
def index_of_element(lst, element):
    return lst.index(element) if element in lst else -1

# Program 14: Check for Empty List
def is_empty_list(lst):
    return len(lst) == 0

# Program 15: Count Even Numbers
def count_even_numbers(lst):
    return len([num for num in lst if num % 2 == 0])

# Program 16: Count Odd Numbers
def count_odd_numbers(lst):
    return len([num for num in lst if num % 2 != 0])

# Program 17: Concatenate Lists
def concatenate_lists(lst1, lst2):
    return lst1 + lst2

# Program 18: Find Sublist
def find_sublist(lst, sublist):
    sub_len = len(sublist)
    return any(lst[i:i + sub_len] == sublist for i in range(len(lst) - sub_len + 1))

# Program 19: Replace Element
def replace_element(lst, old, new):
    if old in lst:
        index = lst.index(old)
        lst[index] = new
    return lst

# Program 20: Find Second Largest
def second_largest(lst):
    unique_lst = list(set(lst))
    unique_lst.sort(reverse=True)
    return unique_lst[1] if len(unique_lst) > 1 else None

# Program 21: Find Second Smallest
def second_smallest(lst):
    unique_lst = list(set(lst))
    unique_lst.sort()
    return unique_lst[1] if len(unique_lst) > 1 else None

# Program 22: Filter Even Numbers
def filter_even_numbers(lst):
    return [num for num in lst if num % 2 == 0]

# Program 23: Filter Odd Numbers
def filter_odd_numbers(lst):
    return [num for num in lst if num % 2 != 0]

# Program 24: List Length
def list_length(lst):
    return len(lst)

# Program 25: Create a Copy
def create_copy(lst):
    return lst[:]

# Program 26: Get Middle Element
def get_middle_element(lst):
    n = len(lst)
    if n == 0:
        return None
    elif n % 2 == 0:
        return lst[n // 2 - 1:n // 2 + 1]
    else:
        return lst[n // 2]

# Program 27: Find Maximum of Sublist
def max_of_sublist(lst, start, end):
    return max(lst[start:end + 1])

# Program 28: Find Minimum of Sublist
def min_of_sublist(lst, start, end):
    return min(lst[start:end + 1])

# Program 29: Remove Element by Index
def remove_element_by_index(lst, index):
    if 0 <= index < len(lst):
        lst.pop(index)
    return lst

# Program 30: Check if List is Sorted
def is_sorted(lst):
    return lst == sorted(lst)

# Program 31: Repeat Elements
def repeat_elements(lst, times):
    return [element for element in lst for _ in range(times)]

# Program 32: Merge and Sort
def merge_and_sort(lst1, lst2):
    return sorted(lst1 + lst2)

# Program 33: Find All Indices
def find_all_indices(lst, element):
    return [i for i, x in enumerate(lst) if x == element]

# Program 34: Rotate List
def rotate_list(lst, steps):
    steps %= len(lst)
    return lst[-steps:] + lst[:-steps]

# Program 35: Create Range List
def create_range_list(start, end):
    return list(range(start, end + 1))

# Program 36: Sum of Positive Numbers
def sum_positive_numbers(lst):
    return sum(num for num in lst if num > 0)

# Program 37: Sum of Negative Numbers
def sum_negative_numbers(lst):
    return sum(num for num in lst if num < 0)

# Program 38: Check Palindrome
def check_palindrome(lst):
    return lst == lst[::-1]

# Program 39: Create Nested List
def create_nested_list(lst, n):
    return [lst[i:i + n] for i in range(0, len(lst), n)]

# Program 40: Get Unique Elements in Order
def get_unique_elements_in_order(lst):
    seen = set()
    return [x for x in lst if not (x in seen or seen.add(x))]

# Example usage:
# Uncomment the following lines to test the functions
# print(count_occurrences([1, 2, 2, 3], 2))
# print(sum_of_elements([1, 2, 3, 4]))
# print(max_element([1, 2, 3, 4]))
# print(min_element([1, 2, 3, 4]))
```
