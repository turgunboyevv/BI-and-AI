```python
import numpy as np

# Task 1: Create a vector with values ranging from 10 to 49
vector = np.arange(10, 50)
print("Task 1: Vector from 10 to 49:")
print(vector)

# Task 2: Create a 3x3 matrix with values ranging from 0 to 8
matrix_3x3 = np.arange(0, 9).reshape(3, 3)
print("\nTask 2: 3x3 Matrix with values from 0 to 8:")
print(matrix_3x3)

# Task 3: Create a 3x3 identity matrix
identity_matrix = np.eye(3)
print("\nTask 3: 3x3 Identity Matrix:")
print(identity_matrix)

# Task 4: Create a 3x3x3 array with random values
random_array = np.random.rand(3, 3, 3)
print("\nTask 4: 3x3x3 Random Array:")
print(random_array)

# Task 5: Create a 10x10 array with random values and find the minimum and maximum values
random_10x10 = np.random.rand(10, 10)
min_val = random_10x10.min()
max_val = random_10x10.max()
print("\nTask 5: 10x10 Random Array - Min and Max Values:")
print(f"Min value: {min_val}, Max value: {max_val}")

# Task 6: Create a random vector of size 30 and find the mean value
random_vector = np.random.rand(30)
mean_value = random_vector.mean()
print("\nTask 6: Random Vector of Size 30 - Mean Value:")
print(f"Mean value: {mean_value}")

# Task 7: Normalize a 5x5 random matrix
random_5x5 = np.random.rand(5, 5)
normalized_matrix = (random_5x5 - random_5x5.min()) / (random_5x5.max() - random_5x5.min())
print("\nTask 7: Normalize 5x5 Random Matrix:")
print(normalized_matrix)

# Task 8: Multiply a 5x3 matrix by a 3x2 matrix (real matrix product)
matrix_5x3 = np.random.rand(5, 3)
matrix_3x2 = np.random.rand(3, 2)
result_product = np.dot(matrix_5x3, matrix_3x2)
print("\nTask 8: Matrix Multiplication (5x3) by (3x2):")
print(result_product)

# Task 9: Create two 3x3 matrices and compute their dot product
matrix_A = np.random.rand(3, 3)
matrix_B = np.random.rand(3, 3)
dot_product = np.dot(matrix_A, matrix_B)
print("\nTask 9: Dot Product of Two 3x3 Matrices:")
print(dot_product)

# Task 10: Given a 4x4 matrix, find its transpose
matrix_4x4 = np.random.rand(4, 4)
transpose_matrix = matrix_4x4.T
print("\nTask 10: Transpose of 4x4 Matrix:")
print(transpose_matrix)

# Task 11: Create a 3x3 matrix and calculate its determinant
matrix_3x3 = np.random.rand(3, 3)
determinant = np.linalg.det(matrix_3x3)
print("\nTask 11: Determinant of 3x3 Matrix:")
print(f"Determinant: {determinant}")

# Task 12: Create two matrices (A of size 3x4) and (B of size 4x3), and compute the matrix product (A ⋅ B)
matrix_A = np.random.rand(3, 4)
matrix_B = np.random.rand(4, 3)
matrix_product = np.dot(matrix_A, matrix_B)
print("\nTask 12: Matrix Product (3x4) ⋅ (4x3):")
print(matrix_product)

# Task 13: Create a 3x3 random matrix and a 3-element column vector. Compute the matrix-vector product
random_matrix_3x3 = np.random.rand(3, 3)
column_vector = np.random.rand(3, 1)
matrix_vector_product = np.dot(random_matrix_3x3, column_vector)
print("\nTask 13: Matrix-Vector Product (3x3) ⋅ (3x1):")
print(matrix_vector_product)

# Task 14: Solve the linear system (Ax = b), where (A) is a 3x3 matrix, and (b) is a 3x1 column vector
A = np.random.rand(3, 3)
b = np.random.rand(3, 1)
x = np.linalg.solve(A, b)
print("\nTask 14: Solve the Linear System Ax = b:")
print(x)

# Task 15: Given a 5x5 matrix, find the row-wise and column-wise sums
matrix_5x5 = np.random.rand(5, 5)
row_sums = matrix_5x5.sum(axis=1)
column_sums = matrix_5x5.sum(axis=0)
print("\nTask 15: Row-wise and Column-wise Sums of 5x5 Matrix:")
print(f"Row sums: {row_sums}")
print(f"Column sums: {column_sums}")
```
