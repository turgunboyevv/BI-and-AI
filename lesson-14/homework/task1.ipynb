Scrape weather information from an HTML file and process it using Python and BeautifulSoup.

Friday	30°C	Sunny
Assume you are given the following HTML structure (you can save it as weather.html):

<!DOCTYPE html>
<html>
<head>
    <title>Weather Forecast</title>
</head>
<body>
    <h4>5-Day Weather Forecast</h4>
    <table>
        <thead>
            <tr>

                <th>Temperature</th>
                <th>Condition</th>
            </tr>
        </thead>
        <tbody>
            <tr>
                <td>Monday</td>
                <td>25°C</td>
                <td>Sunny</td>
            </tr>

                <td>22°C</td>

                <td>Rainy</td>
            </tr>

                <td>20°C</td>
                <td>Partly Cloudy</td>
            </tr>

Parse the HTML File:

Load the weather.html file using BeautifulSoup and extract the weather forecast details.
Display Weather Data:

Print the day, temperature, and condition for each entry in the forecast.
Find Specific Data:

Identify and print the day(s) with:
The highest temperature.
The "Sunny" condition.
Calculate Average Temperature:

Compute and print the average temperature for the week.

```python
from bs4 import BeautifulSoup
import re

# Load the HTML file
html_content = """<!DOCTYPE html>
<html>
<head>
    <title>Weather Forecast</title>
</head>
<body>
    <h4>5-Day Weather Forecast</h4>
    <table>
        <thead>
            <tr>
                <th>Day</th>
                <th>Temperature</th>
                <th>Condition</th>
            </tr>
        </thead>
        <tbody>
            <tr>
                <td>Monday</td>
                <td>25°C</td>
                <td>Sunny</td>
            </tr>
            <tr>
                <td>Tuesday</td>
                <td>22°C</td>
                <td>Cloudy</td>
            </tr>
            <tr>
                <td>Wednesday</td>
                <td>18°C</td>
                <td>Rainy</td>
            </tr>
            <tr>
                <td>Thursday</td>
                <td>20°C</td>
                <td>Partly Cloudy</td>
            </tr>
            <tr>
                <td>Friday</td>
                <td>30°C</td>
                <td>Sunny</td>
            </tr>
        </tbody>
    </table>

</body>
</html>"""

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Extract the weather forecast data from the table
weather_data = []
table_rows = soup.find_all("tr")[1:]  # Skip the header row
for row in table_rows:
    cells = row.find_all("td")
    day = cells[0].text.strip()
    temp = int(re.search(r'\d+', cells[1].text).group())  # Extract temperature as an integer
    condition = cells[2].text.strip()
    weather_data.append({"day": day, "temperature": temp, "condition": condition})

# Display the weather data
print("Weather Forecast:")
for entry in weather_data:
    print(f"Day: {entry['day']}, Temperature: {entry['temperature']}\u00b0C, Condition: {entry['condition']}")

# Find the day with the highest temperature
max_temp = max(weather_data, key=lambda x: x['temperature'])
print(f"\nDay with the highest temperature: {max_temp['day']} ({max_temp['temperature']}\u00b0C)")

# Find all days with the "Sunny" condition
sunny_days = [entry['day'] for entry in weather_data if entry['condition'] == "Sunny"]
print(f"Days with Sunny condition: {', '.join(sunny_days)}")

# Calculate the average temperature
average_temp = sum(entry['temperature'] for entry in weather_data) / len(weather_data)
print(f"\nAverage temperature for the week: {average_temp:.2f}\u00b0C")
```
